```python
import pandas as pd
# allows sending http requests.
# HTTP request is made by client to named host, which is located on server.
# Aim is requesting access a resource on the server.(use of URL)
import requests
# for pulling data out of HTML and XML
# for parsing html and xml documents
from bs4 import BeautifulSoup 
```

```python
Movie_Name = []
Genre = []
Year = []
Time = []
Movie_Rating = []
Movie_Votes = []
Rank = []
Gross = []
Movie_Director = []
Stars = []
Overview = []
```

```python
pages = ["1"]
```

```python
for page in range(1):
    url = "https://www.imdb.com/search/title/?count=100&title_type=feature,tv_series&ref_=nv_wl_img_2"
    response = requests.get(url) # get will returns the value of the item.
    soup = BeautifulSoup(response.content, "html.parser")
    movie_data = soup.findAll("div", attrs = {'class':'lister-item mode-advanced'})
    # findall is used to search for “all” occurrences that match a given pattern
    # find is used to search for the first occurence of the specified value
```

```python
for row in movie_data:
    Movie_Name.append(row.h3.a.text)
Movie_Name[:101]
```

```python
for row in movie_data:
    Year.append(row.h3.find('span', class_ = 'lister-item-year text-muted unbold').text)
Year[:101]
# class - class is reserved for language that's why it will give error.
# class_ - because we are using bs4 we have to use class_. It will search by CSS class using the class_ keyword argument.
```

```python
for row in movie_data:
    Genre.append(row.p.find('span', class_ = "genre").text[1:])
Genre[:101]
```

```python
for row in movie_data:
    Time.append(row.p.find('span', class_ = 'runtime').text[:-4] if row.p.find('span', class_ = 'runtime') is not None else "" )
Time[:101]
```

```python
for row in movie_data:
    if row.p.find('span', class_ = 'runtime') is not None:
        Time.append(row.p.find('span', class_ = 'runtime').text[:-4])
    else:
        Time.append("")
Time[:101]
```

```python
for row in movie_data:
    if row.strong is not None:
        Movie_Rating.append(row.strong.text)
    else:
        Movie_Rating.append("")
Movie_Rating[:101]
```

```python
for row in movie_data:
    if row.p.find('span', class_ = 'sort-num_votes-visible') is not None:
        Movie_Rating.append(row.span.find.text)
    else:
        Movie_Rating.append("")
Movie_Rating[:101]
```

```python
for row in movie_data:
    Movie_Director.append(row.find('a', attrs={'class':''}).text)
Movie_Director[:6]
```

```python
for row in movie_data:
    if row.p.find('span',attrs={'name':'nv'}) is not None:
        Gross.append(row.span.find.text)
    else:
        Gross.append("")
Gross[:101]
```
